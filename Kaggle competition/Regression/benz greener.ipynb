```python
train = pd.read_csv("/kaggle/input/mercedes-benz-greener-manufacturing/train.csv.zip")
train.head()
```

```python
test = pd.read_csv("/kaggle/input/mercedes-benz-greener-manufacturing/test.csv.zip")
test.head()
```

```python
alldata = pd.concat([train,test])
alldata
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in alldata.columns[alldata.dtypes == object]:
    alldata[i] = le.fit_transform(list(alldata[i]))
    
alldata
```

```python
alldata2 = alldata.drop(["y"], axis=1)
alldata2
```

```python
train2 = alldata2[:len(train)]
test2 = alldata2[len(train):]
```

```python
from catboost import CatBoostRegressor
# cat = CatBoostRegressor(n_estimators = 20000, max_depth = 10, thread_count = 6, colsample_bylevel = 0.9, subsample = 0.8, learning_rate = 0.02) 
cat = CatBoostRegressor(n_estimators = 1250, max_depth = 2, task_type = "GPU")

cat.fit(train2,train["y"])
result = cat.predict(test2)
```

```python
sub = pd.read_csv("/kaggle/input/mercedes-benz-greener-manufacturing/sample_submission.csv.zip")
sub.head()
```

```python
sub["y"]= result
sub.head()
```

```python
sub.to_csv("sub.csv", index = False)
```
