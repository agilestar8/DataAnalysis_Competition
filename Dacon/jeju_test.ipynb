```python
train = pd.read_csv("/kaggle/input/jeju-dataset/201901-202003.csv")
sub = pd.read_csv("/kaggle/input/jeju-dataset/submission.csv")

display(train.head(),sub.head())
```

```python
alldata = pd.concat([train,sub])

alldata["year"] = np.floor(alldata["REG_YYMM"]/100).astype(int)
alldata["month"] = np.floor(alldata["REG_YYMM"]%100).astype(int)
alldata = alldata.drop(["REG_YYMM","id"], axis = 1)

alldata
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in alldata.columns[alldata.dtypes == object]:
    
    alldata[i] = le.fit_transform(list(alldata[i]))

alldata
```

```python
# 매월 지역별 업종의 총 이용금액
month_AMT = alldata.groupby(["year","month",'CARD_SIDO_NM', 'STD_CLSS_NM'])["AMT"].sum().reset_index()
month_AMT
```

```python
# 데이터 재구성, 모든 경우
from itertools import product

matrix = []

matrix.append(np.array(list(product(alldata["year"].unique(), alldata["month"].unique(),alldata["CARD_SIDO_NM"].unique(), alldata["STD_CLSS_NM"].unique() )), dtype = np.int16)) 

matrix = pd.DataFrame(np.vstack(matrix))
matrix.columns = ["year","month","CARD_SIDO_NM","STD_CLSS_NM"]

#  ~ 2020.04월,07월만
matrix = matrix.drop(matrix[(matrix["year"]==2020) & (matrix["month"] > 7)].index)
matrix = matrix.drop(matrix[(matrix["year"]==2020) & (matrix["month"] == 5)].index)
matrix = matrix.drop(matrix[(matrix["year"]==2020) & (matrix["month"] == 6)].index)

# matrix = matrix.reset_index().drop(["index"], axis=1)
matrix
```

```python
matrix2 = pd.merge(matrix,month_AMT,on=["year","month","CARD_SIDO_NM","STD_CLSS_NM"], how="left")
matrix2
```

```python
# 결측치 처리 - 판매량 0으로 채움
matrix2 = matrix2.fillna(0)

matrix2[(matrix2["year"] == 2020) & (matrix2["month"] == 3) & (matrix2["CARD_SIDO_NM"] == 16)].tail(10)
# -> 업종 4,12,14,15,23,28,30의 7개 업종은 매월 사용량 발생이 적음
```

```python
train2 = matrix2.iloc[:10455,]
test2 = matrix2.iloc[10455:,].drop(["AMT"], axis = 1)

train3 = train2.drop(["AMT"],axis=1)
```

```python
from catboost import CatBoostRegressor
cat = CatBoostRegressor(iterations = 1000)
cat.fit(train3,train2["AMT"])
result = cat.predict(test2)

# from lightgbm import LGBMRegressor
# lgb = CatBoostRegressor(n_estimators = 300)
# lgb.fit(x_train,y_train)
# result = lgb.predict(x_valid)
```
